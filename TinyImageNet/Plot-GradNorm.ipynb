```
import matplotlib.pyplot as plt
import numpy as np

fid1 = open('./CheckpointsResults/TinyImageNet_VGG16_Adam_GradNorm.txt', 'r')
Lines = fid1.readlines()
ypoints1 = []
for line in Lines:
    line1 = line.split()
    #print(len(line1))
    line1.reverse()
    line1 = [float(i) for i in line1]
    #ypoints1.append(line1[layer])
    ypoints1.append(sum(line1)/len(line1))
plt.plot(ypoints1)
fid1.close()
#print(len(ypoints1))

fid2 = open('./CheckpointsResults/TinyImageNet_VGG16_AdamNorm95_GradNorm1.txt', 'r')
Lines = fid2.readlines()
ypoints2 = []
for line in Lines:
    line2 = line.split()
    #print(len(line1))
    line2.reverse()
    line2 = [float(i) for i in line2]
    #ypoints1.append(line1[layer])
    ypoints2.append(sum(line2)/len(line2))
plt.plot(ypoints2)
fid2.close()
#print(len(ypoints1))

plt.legend(["Adam", "AdamNorm"], loc ="upper right")
plt.yticks(np.arange(0.0, 5.2, 0.5))
plt.show()
# pause
```
